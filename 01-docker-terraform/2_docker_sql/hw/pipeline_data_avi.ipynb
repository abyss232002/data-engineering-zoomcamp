```python
import time as t
flag = True
rec_cnt = 0
while flag:
    try:
        t_start = t.time()
        df = next(df_iter)
        rec_cnt = rec_cnt + len(df)
        df.tpep_dropoff_datetime = pd.to_datetime(df['tpep_dropoff_datetime'])
        df.tpep_pickup_datetime = pd.to_datetime(df['tpep_pickup_datetime'])
        df.to_sql('yellow_taxi_data', engine, if_exists='append')
        t_end = t.time()
        print(f"Time taken to insert 10000 rows: {t_end-t_start}")
        print(f"Total records inserted: {rec_cnt}")
    except Exception as e:
        flag = False
        print(f"Exception encounterd: {e}")
print("Data insertion completed with {rec_cnt} records")
```
